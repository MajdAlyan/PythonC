```python
import matplotlib.pyplot as plt 
import numpy as np 
import cv2
```

```python
plt.rcParams["figure.figsize"] =[7.00, 3.50]
plt.rcParams["figure.autolayout"] = True
x = [4]
y = [3]
```

```python
plt.xlim(0,5)
plt.ylim(0,5)
plt.grid()
plt.plot(x, y, marker="o", markersize=20, markeredgecolor="red", markerfacecolor="green")
plt.show()
```

```python
x = [1,2,3]
y = np.array([[1,2],[3,4],[5,6]])
plt.plot(x,y)
```

```python
black_img = np.zeros(shape=(512,512,3),
                    dtype=np.int16)
```

```python
plt.imshow(black_img)
```

```python
circle = cv2.circle(img=black_img,
          center=(400,100),
          radius=50,
          color=(255,0,0),
          thickness=8)

plt.imshow(circle)
```
